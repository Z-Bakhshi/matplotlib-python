## **آموزش Matplotlib**

🧠 `Matplotlib` یک کتابخانه گرافیکی برای زبان برنامه‌نویسی پایتونه که امکان رسم انواع نمودارها و چارت‌ها را فراهم می‌کنه. این کتابخانه بسیار قدرتمند و انعطاف‌پذیره و به عنوان یکی از ابزارهای اصلی برای تصویرسازی داده در پایتون شناخته می‌شه.

معمولاً ابتدا با دستور `import matplotlib.pyplot as plt` ماژول `pyplot` از `Matplotlib` را وارد می‌کنیم. سپس با استفاده از دستورات مختلف مانند `plot`، `scatter`، `bar` و غیره، می‌توانیم نمودارها و چارت‌های مختلفی را ایجاد کنیم.

برای مثال، اگر می‌خواهید یک نمودار خطی ایجاد کنید، می‌توانید از دستور `plt.plot(x, y)` استفاده کنید که `x` و `y` به ترتیب داده‌های محور افقی و عمودی هستند.

همچنین، `Matplotlib` امکانات زیادی برای تنظیمات سفارشی نمودارها، افزودن عناصر مختلف مانند عناصر متنی و اطلاعات لیبل، انجام تغییرات در رنگ‌ها و استایل‌ها، و غیره را فراهم می‌کنه.

## **Matplotlib Plotting (رسم نمودار)**

تابع plot() برای رسم نقاط (نشانگرها) در یک نمودار استفاده می شود.

به طور پیش فرض، تابع plot() یک خط از نقطه ای به نقطه دیگر ترسیم می کند.

تابع پارامترهایی را برای تعیین نقاط در نمودار می گیرد.

پارامتر 1 آرایه ای است که نقاط روی محور x را در بر می گیرد.

پارامتر 2 آرایه ای است که شامل نقاط روی محور y است.

اگر نیاز به رسم خطی از (1، 3) تا (8، 10) داشته باشیم، باید دو آرایه [1، 8] و [3، 10] را به تابع نمودار ارسال کنیم.

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

⚠️ محور x محور افقی است.

⚠️ محور y محور عمودی است.

برای ترسیم فقط نشانگرها، می توانید از پارامتر علامت گذاری رشته میانبر 'o' که به معنای حلقه ها است استفاده کنید.

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

می توانید هر تعداد نقطه را که دوست دارید رسم کنید، فقط مطمئن شوید که تعداد نقاط در هر دو محور یکسان است.

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

اگر نقاط روی محور x را مشخص نکنیم، بسته به طول نقاط y، مقادیر پیش‌فرض 0، 1، 2، 3 و غیره را دریافت خواهند کرد.

بنابراین، اگر همان مثال بالا را در نظر بگیریم و نقاط x را کنار بگذاریم، نمودار به شکل زیر خواهد بود:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

نقاط x در مثال بالا ☝🏽 [0، 1، 2، 3، 4، 5] هستند.

</section>

## **Matplotlib Markers (نشانگرها)**

می توانید از نشانگر آرگومان کلمه کلیدی برای تأکید بر هر نقطه با یک نشانگر مشخص استفاده کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o')
plt.show()
```

```python
plt.plot(ypoints, marker='*')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

همچنین می توانید از پارامتر علامت گذاری رشته میانبر برای تعیین نشانگر استفاده کنید.

این پارامتر fmt نیز نامیده می شود و با این نحو نوشته می شود:

marker|line|color

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

مقدار نشانگر می تواند هر چیزی از مرجع نشانگر بالا باشد.

⚠️ نکته: اگر مقدار خط را در پارامتر fmt کنار بگذارید، هیچ خطی رسم نخواهد شد.

برای تنظیم اندازه نشانگرها می توانید از کلمه کلیدی آرگومان markersize یا نسخه کوتاهتر ms استفاده کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o', ms=20)
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

برای تنظیم رنگ لبه نشانگرها می توانید از آرگومان کلمه کلیدی markeredgecolor یا mec کوتاهتر استفاده کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o', ms='20', mec='r')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

برای تنظیم رنگ در لبه نشانگرها می توانید از آرگومان کلمه کلیدی markerfacecolor یا mfc کوتاهتر استفاده کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o', ms=20, mfc='r')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

از هر دو آرگومان mec و mfc برای رنگ آمیزی کل نشانگر استفاده کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o', ms=20, mec='r', mfc='r')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

همچنین می توانید از مقادیر رنگ هگزادسیمال استفاده کنید:

</section>

```python
plt.plot(ypoints, marker='o', ms=20, mec='#4CAF50', mfc='#4CAF50')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

یا هر یک از 140 نام رنگ پشتیبانی شده.

</section>

```python
plt.plot(ypoints, marker='o', ms=20, mec='hotpink', mfc='hotpink')
plt.show()
```

## **Matplotlib Line (خط)**

این نمودارها ابزار بصری سازی قدرتمندی هستند که روند تغییرات داده‌ها را در دوره‌ی زمانی یا میزان همبستگی خاصی نشان می‌دهند. مثلاً یک محور ممکن است یک مقدار متغیر را نشان دهد درحالی‌که محور دیگر تعیین‌کننده بازه زمانی باشد. ابتدا هر مقدار به‌طور مجزا مشخص می‌شود و در آخر نقاط به هم متصل می‌گردند؛ حتی بیش از یک خط هم برای نمایش نمودارهای مختلف بارنگ‌های متفاوت قابل‌نمایش است. به‌طور مثال در نمودار بالا معلوم است که کدام شبکه‌های اجتماعی از بقیه موفق‌تر عمل کردند و همچنین روند رشد و شکست هریک به‌وضوح نمایان است.

می توانید از کلمه کلیدی آرگومان linestyle یا ls کوتاهتر برای تغییر سبک خط رسم شده استفاده کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle='dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle='dashed')
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

سبک خط را می توان با یک نحو کوتاهتر نوشت:

- linestyle را می توان به صورت ls نوشت.

- dotted را می‌توان به صورت زیر نوشت:

- dashed را می توان به صورت -- نوشت.

</section>

```python
plt.plot(ypoints, ls=':')
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color='r')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

همچنین می توانید از مقادیر رنگ هگزادسیمال استفاده کنید:

</section>

```python
plt.plot(ypoints, c = '#4CAF50')
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

یا هر یک از 140 نام رنگ پشتیبانی شده.

</section>

```python
plt.plot(ypoints, c = 'hotpink')
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

می توانید از کلمه کلیدی آرگومان linewidth یا lw کوتاهتر برای تغییر عرض خط استفاده کنید.

مقدار یک عدد شناور است، در نقاط:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth='20.5')
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

شما می توانید به سادگی با افزودن توابع ()plt.plot بیشتر به تعداد خطوطی که دوست دارید رسم کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

همچنین می توانید خطوط زیادی را با اضافه کردن نقاط محور x و y برای هر خط در همان تابع ()plt.plot رسم کنید.

(در مثال های بالا ما فقط نقاط روی محور y را مشخص کردیم، به این معنی که نقاط روی محور x مقادیر پیش فرض (0، 1، 2، 3) را دریافت کردند.)

مقادیر x و y به صورت جفت می آیند:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

## **برچسب‌ها و عناوین**

با Pyplot می توانید از توابع xlabel() و ylabel() برای تعیین برچسب برای محور x و y استفاده کنید.

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

با Pyplot می توانید از تابع title() برای تعیین عنوان برای طرح استفاده کنید.

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

می‌توانید از پارامتر fontdict در xlabel()، ylabel() و title() برای تنظیم ویژگی‌های فونت برای عنوان و برچسب‌ها استفاده کنید.

</section>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

می توانید از پارامتر loc در title() برای قرار دادن عنوان استفاده کنید.

مقادیر قانونی عبارتند از: «left»، «right» و «center». مقدار پیش فرض "center" است.

</section>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

## **خطوط شبکه‌ای**

با Pyplot می توانید از تابع grid() برای اضافه کردن خطوط شبکه به نمودار استفاده کنید.

</section>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

می توانید از پارامتر محور در تابع grid() استفاده کنید تا مشخص کنید کدام خطوط شبکه نمایش داده شود.

مقادیر قانونی عبارتند از: «x»، «y» و «هر دو». مقدار پیش فرض "هر دو" است.

</section>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

plt.plot(x, y)

plt.grid(axis='x')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

plt.plot(x, y)

plt.grid(axis='y')

plt.show()
```

</section>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

## **نمودار فرعی**

با تابع subplot() می توانید چندین نمودار را در یک شکل رسم کنید:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

تابع subplot () سه آرگومان می گیرد که طرح شکل را توصیف می کند.

طرح بندی در ردیف ها و ستون ها سازماندهی شده است که با آرگومان اول و دوم نشان داده می شوند.

آرگومان سوم نمایانگر نمودار فعلی است.

بنابراین، اگر شکلی با 2 ردیف و 1 ستون بخواهیم (به این معنی که این دو نمودار به جای کنار هم نمایش داده می شوند)، می توانیم نحو را به این صورت بنویسیم:

</section>

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

شما می توانید هر تعداد نمودار را که دوست دارید روی یک شکل بکشید، فقط تعداد سطرها، ستون ها و شاخص نمودار را مشخص کنید.

رسم نقاط و خطوط سه بعدی در پایتون

پایه‌ای‌ترین نمودار سه بعدی خط یا مجموعه‌ای از نقاط پراکنده است که با سه تایی (x,y,z) ایجاد می‌شود. این کار را با توابع ax.plot3D و ax.scatter3D انجام می‌دهیم. با کد زیر یک مارپیچ مثلثاتی (spiral) را به نقاطی که به طور تصادفی اطراف خط آن هستند، رسم می‌کنیم.

</section>

```python
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

توجه کنید که به طور پیش فرض ، نقاط پراکندگی شفافیت خود را تنظیم می‌کنند تا حس عمق صفحه را نشان دهند.

<section style="direction: rtl; text-align: justify; margin: 30px;">

رسم نمودارهای کانتور سه بعدی در پایتون


در مثال زیر یک نمودار کانتور سه بعدی از یک تابع سینوسی سه بعدی را نشان خواهیم داد.

</section>

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

گاهی زاویه‌ی  دید نمودار مناسب نیست. می‌توان با کمک تابع view_init زوایای ارتفاع و آزیموت نمودار را تنظیم کرد. در مثال زیر از ارتفاع 60 درجه (یعنی 60 درجه بالای صفحه‌ی x-y) و یک آزیموت 35 درجه (یعنی 35 درجه خلاف جهت عقربه‌های ساعت در مورد محور z می‌چرخیم) استفاده خواهیم کرد:

</section>

```python
ax.view_init(60, 35)
fig
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

توجه داشته باشید که این نوع چرخش با استفاده از دکمه‌های تعاملی Matplotlib می‌تواند به صورت تعاملی با کلیک و کشیدن انجام شود.

</section>

قاب سیمی و پوسته‌ی سه بعدی در پایتون


دو نوع دیگر از اشکال سه بعدی که با داده‌های شبکه‌ای (grid) سر و کار دارند، قاب سیمی  (wire frame) و پوسته (surface) نام دارند. این نمودارها شبکه‌ی داده‌ها را به شکل پوسته‌ای سه بعدی نمایش می‌دهند. در کد زیر مثالی از این نوع داده‌ها و ترسیمات را می‌بینیم:

</section>

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe');
```

<section style="direction: rtl; text-align: justify; margin: 30px;">

طرح پوسته نیز مانند قاب سیمی است با این تفاوت که هر وجه آن با یک چند ضلعی پر شده است. افزودن یک طرح زنگی به هر وجه می‌تواند به تجسم و درک بهتر توپولوژی پوسته کمک کند.

</section>

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

نمودار ویولن در Matplotlib

نمودار ویولن شبیه طرح های جعبه ای است با این تفاوت که چگالی احتمال داده ها را نیز در مقادیر مختلف نشان می دهد. این نمودارها همانند نمودارهای استاندارد جعبه، شامل یک نشانگر برای میانه داده ها و یک جعبه نشان دهنده دامنه بین چهارقسمی است. برآورد شده روی این طرح جعبه برآورد تراکم هسته است. همانند نمودارهای جعبه ای ، از طرح های ویولن برای نشان دادن مقایسه یک توزیع متغیر (یا توزیع نمونه) در “دسته بندی” های مختلف استفاده می شود.

یک نمودار ویولون از یک طرح جعبه ساده بسیار آموزنده است. در حقیقت در حالی که یک طرح جعبه فقط آمار خلاصه ای از جمله میانگین / متوسط و دامنه های بین ذره ای را نشان می دهد ، نمودار ویولن توزیع کامل داده ها را نشان می دهد.

</section>

```python
import matplotlib.pyplot as plt

np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data_to_plot)
plt.show()
```

</section>

<section style="direction: rtl; text-align: justify; margin: 30px;">

مولفه های نمودار quiver در Matplotlib 
جدول زیر پارامترهای مختلف نمودار Quiver را نشان می دهد –

x	 ۱D یا ۲D آرایه ، توالی. مختصات x مکانهای پیکان

y	آرایه ۱D یا ۲D ، توالی. مختصات y مکان های پیکان 

u	توالی آرایه ۱D یا ۲D. اجزای x بردارهای پیکان

v	 آرایه ۱D یا ۲D ، توالی. اجزای y بردارهای پیکان

c	آرایه ۱D یا ۲D ، دنباله. پیکان رنگ می شود

کد زیر یک طرح ساده نمودار quiver را ترسیم می کند –

</section>

```python
import matplotlib.pyplot as plt
import numpy as np
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2)
fig, ax = plt.subplots()
q = ax.quiver(x,y,u,v)
plt.show()
```
